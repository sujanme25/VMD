```python
from vmdpy import VMD 
# For illustration, let's generate synthetic stock price data
np.random.seed(42)
signal = np.cumsum(np.random.normal(0, 1, size=1000))

u, u_hat, omega=VMD(signal, K=5, alpha=200, tau=0.01, tol = 1e-7,DC=0,init=1)

# Assuming 'stock_prices' is your time series data
modes = u.copy()
U=modes.T
# Plot the original signal and its decomposed modes
plt.figure(figsize=(12, 6))
plt.plot(signal, label='Original Signal', alpha=0.7)
for i in range(U.shape[1]):
    plt.plot(U[:, i], label=f'U {i + 1}', alpha=0.7)

plt.title('VMD Decomposition of Stock Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
